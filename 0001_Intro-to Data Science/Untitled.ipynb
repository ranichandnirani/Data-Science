# What is Data Science?

Welcome to this course on **Data Science**! 🚀

Data Science is one of the most in-demand skills today because almost every industry generates huge amounts of data. Organizations use this data to make smarter decisions, improve customer experiences, and solve real-world problems.

---

# Definition of Data Science

## Book Definition (Exam Definition)

**Data Science** is the field of study that uses **mathematics, statistics, programming, and domain knowledge** to extract meaningful insights from data.

---

## Simple Definition

> **Data Science is the process of collecting, cleaning, analyzing, and interpreting data to make informed decisions.**

In simple words, Data Science helps us convert **raw data into useful information** that businesses and organizations can use to solve problems and make better decisions.

# Key Steps in Data Science

## 1. Data Collection

The first step is collecting raw data from different sources.

### Sources of Data

- Websites
- Databases
- APIs
- Mobile applications
- IoT sensors
- Excel files
- CSV files

**Example:**

Amazon collects:

- Customer purchases
- Product reviews
- Search history
- Wishlist items

---

## 2. Data Cleaning

Raw data is rarely perfect.

It may contain:

- Missing values
- Duplicate records
- Incorrect information
- Typing mistakes
- Invalid data

Data cleaning means fixing these problems before analysis.

> **Interesting Fact:** Around **80% of a Data Scientist's time** is spent cleaning data.

**Example:**

| Raw Data | Clean Data |
|----------|------------|
| Age = -5 | Remove record |
| Name = "" | Fill missing value |
| Duplicate rows | Delete duplicates |

---

## 3. Data Analysis

Once the data is clean, it is analyzed to discover patterns and trends.

Data analysts use:

- Statistics
- Graphs
- Charts
- Correlation analysis
- Exploratory Data Analysis (EDA)

**Example Questions:**

- Which product sells the most?
- Which month has the highest sales?
- Which customers spend the most money?

---

## 4. Model Building

In this stage, Machine Learning algorithms are used to make predictions.

Examples:

- Predict house prices
- Predict customer churn
- Predict stock prices
- Spam email detection
- Disease prediction

---

## 5. Interpretation & Communication

The final step is presenting insights in a simple and understandable way.

Common tools include:

- Dashboards
- Charts
- Reports
- Presentations

The goal is to help decision-makers understand the results and take appropriate actions.
